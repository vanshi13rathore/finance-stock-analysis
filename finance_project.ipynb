```python
import yfinance as yf

# turn off caching issue
yf.set_tz_cache_location(None)

banks = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']

df = yf.download(
    banks,
    start='2006-01-01',
    end='2016-01-01',
    auto_adjust=False,
    group_by='ticker',
    threads=False
)

df.head()
```

```python
df.columns
```

```python
close_prices = df.xs('Close', axis=1, level='Price')

close_prices.head()
```

```python
returns = close_prices.pct_change()

returns.head()
```

```python
returns.idxmax()
```

```python
returns.idxmin()
```

```python
returns.std()
```

```python
close_prices.plot(figsize=(12,6))
```

```python
plt.figure(figsize=(12,6))

for bank in close_prices.columns:
    plt.plot(close_prices.index, close_prices[bank], label=bank)

plt.legend()
plt.title("Bank Closing Prices")
plt.xlabel("Year")
plt.ylabel("Closing Price")
plt.show()
```

```python
returns.corr()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(returns.corr(), annot=True)

plt.title("Correlation Between Bank Stocks")
plt.show()
```

```python
returns.corr()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(returns['C'].dropna(), bins=100)

plt.title("Citigroup Daily Returns")
plt.show()
```

```python
returns.hist(figsize=(12,8), bins=50)
plt.tight_layout()
```

```python
sns.pairplot(returns.dropna())
```

```python
BAC = df['BAC'].copy()

BAC['30 Day MA'] = BAC['Close'].rolling(window=30).mean()
BAC['50 Day MA'] = BAC['Close'].rolling(window=50).mean()

BAC[['Close', '30 Day MA', '50 Day MA']].plot(figsize=(12,6))
plt.title("BAC Closing Price with Moving Averages")
plt.show()
```

```python
normalized = close_prices / close_prices.iloc[0] * 100

normalized.plot(figsize=(12,6))

plt.title("Normalized Bank Stock Prices")
plt.xlabel("Year")
plt.ylabel("Normalized Price")
plt.show()
```

```python
returns['BAC'].rolling(window=30).std().plot(figsize=(12,6))

plt.title("30-Day Rolling Volatility of BAC")
plt.xlabel("Year")
plt.ylabel("Volatility")
plt.show()
```

```python
cumulative_returns = (
    1 + returns
).cumprod()

cumulative_returns.plot(figsize=(12,6))

plt.title("Cumulative Returns of Bank Stocks")
plt.xlabel("Year")
plt.ylabel("Growth of $1 Investment")
plt.show()
```

```python
insights = pd.DataFrame({
    'Risk (Std Dev)': returns.std(),
    'Average Return': returns.mean(),
    'Max Return': returns.max(),
    'Min Return': returns.min()
})

insights
```
